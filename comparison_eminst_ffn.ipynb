```python
import matplotlib.pyplot as plt
from IPython import display
import numpy as np
import copy
import math
import time
import random
import os

import torch
from torch import nn, optim
import torch.nn.functional as F
from torch.utils.data import DataLoader

from common.optim.meta import MetaData
from common.optim.exp.netline_scheduler_v1 import NetLineStepLR
from common.optim.ffn_minst_optimise import \
    StepCalculatorEtaSoftmaxArmihoNorm2Base, StepCalculatorEtaSoftmaxArmihoNorm1Base
from common.ffn_minst_relu import MNISTReLU
from common.lr_scheduling import cosine_annealing2_lr, cosine_annealing4_lr
from common.util import AverageMeter

from torchvision.datasets import EMNIST
from torchvision.transforms import ToTensor

import logging
```

#### Constants

```python
BATCH_SIZE = 64 #128
SAMPLE_SIZE = 2048
OUTPUT_DIM=26
DEVICE = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
slope_plus, slope_minus=1.0, 0.1

EXPERIMENTS = 3
EPOCHS_PER_EXPERIMENT = 20
DATASET_PATH = "./datasets"

CALC_GRAD_NORM = False
```

#### Manual seed

```python
seed_value= 641

# 1. Set `PYTHONHASHSEED` environment variable at a fixed value
os.environ['PYTHONHASHSEED']=str(seed_value)
# 2. Set `python` built-in pseudo-random generator at a fixed value
random.seed(seed_value)
# 3. Set `numpy` pseudo-random generator at a fixed value
np.random.seed(seed_value)
# 4. Set `pytorch` pseudo-random generator at a fixed value
dummy = torch.manual_seed(seed_value)
```

#### EMNIST dataset

```python
image_transform = ToTensor()

train_dataset = EMNIST(root=DATASET_PATH, split='letters', train=True, download=True, transform=image_transform)
test_dataset = EMNIST(root=DATASET_PATH, split='letters', train=False, download=True, transform=image_transform)

train_dataloader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
)

test_dataloader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
)
```

##### UTILS

```python
def eta_straightline(testNet, loss_value):
    denominator = 0.0
    for param in testNet.parameters():
        denominator += ((param.grad)**2).sum().item()
    eta = loss_value/denominator
    return eta

def calculate_accuracy(prediction, target):
    return np.average((prediction ==target).detach().cpu())

def test_loop(testNet):
    accuracy_meter, loss_meter  = AverageMeter(), AverageMeter(),
    with torch.no_grad():
        for test_batch in test_dataloader:
            images, labels_raw = test_batch
            labels = labels_raw - 1
            images = images.to(DEVICE)
            labels = labels.to(DEVICE)
            batch_size = images.shape[0]
            xx = images.view(batch_size, -1)
            logits = testNet.forward_(xx)
            loss = F.cross_entropy(logits, labels)
            loss_meter.update(loss)
            prediction = logits.argmax(dim=-1)
            accuracy_meter.update(calculate_accuracy(prediction, labels))

    return accuracy_meter.avg, loss_meter.avg
```

```python
def experiment_comparison(data, var, low, high, title="Comparison"):

    MEAN_STRGHT = data[low:high,0]
    MEAN_SGD = data[low:high,1]
    MEAN_NETNTK2 = data[low:high,2]
    MEAN_NETSTEP = data[low:high,3]
    MEAN_NETNTK1 = data[low:high,4]

    display.clear_output()
    fig, axes = plt.subplots(4, 2, figsize=(18, 24)) if CALC_GRAD_NORM else plt.subplots(3, 2, figsize=(18, 18))
    #fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    axes[0,0].set_title('Loss (average cross entropy)')
    axes[0,0].plot(MEAN_NETNTK2[:,1], color='b',alpha=.5, label='Net-line NTK norm-2')
    axes[0,0].plot(MEAN_NETNTK1[:,1], color='g',alpha=.5, label='Net-line NTK norm-1')
    axes[0,0].plot(MEAN_NETSTEP[:,1], color='b', ls='dotted', alpha=.5, label='Net-line 2-step')
    axes[0,0].plot(MEAN_STRGHT[:,1], color='orange', alpha=.5, label='Straight-line')
    axes[0,0].plot(MEAN_SGD[:,1], color='grey', alpha=.5, label='SGD')
    axes[0,0].grid()
    axes[0,0].legend()
    axes[0,0].set_yscale("log")
    axes[0,0].set_xlabel("Epoch (series of gradient descent steps)")
    axes[0,0].set_ylabel("Loss value")

    axes[0,1].set_title('Error (top 1)')
    axes[0,1].plot(1.0-MEAN_NETNTK2[:,0], color='b', alpha=.5, label='Net-line NTK norm-2')
    axes[0,1].plot(1.0-MEAN_NETNTK1[:,0], color='g', alpha=.5, label='Net-line NTK norm-1')
    axes[0,1].plot(1.0-MEAN_NETSTEP[:,0], color='b', ls='dotted', alpha=.5, label='Net-line 2-step')
    axes[0,1].plot(1.0-MEAN_STRGHT[:,0], color='orange', alpha=.5, label='Straight-line')
    axes[0,1].plot(1.0-MEAN_SGD[:,0], color='grey', alpha=.5, label='SGD')
    axes[0,1].grid()
    axes[0,1].set_yscale("log")
    axes[0,1].legend()
    axes[0,1].set_xlabel("Epoch (series of gradient descent steps)")
    axes[0,1].set_ylabel("Error value")

    axes[1,0].set_title('PQ vector norm')
    axes[1,0].plot(MEAN_NETSTEP[:,4], color='grey',alpha=.5, label='PQ vector norm average per epoch')
    axes[1,1].set_title('QQ vector norm')
    axes[1,1].set_yscale("log")
    axes[1,1].plot(MEAN_NETSTEP[:,3], color='grey',alpha=.5, label='QQ vector norm average per epoch')

    axes[2,0].set_title('Cos phi')
    axes[2,0].plot(MEAN_NETSTEP[:,5], color='grey',alpha=.5, label='Cos phi average per epoch')
    axes[2,1].set_title('Eta calculated')
    axes[2,1].set_yscale("log")
    axes[2,1].plot(MEAN_NETSTEP[:,6], color='green',alpha=.5, label='Eta average per epoch')
    axes[2,1].plot(MEAN_SGD[:,6], color='red',alpha=.5, label='Eta SGD average per epoch')
    axes[2,1].grid()
    axes[2,1].legend()

    if CALC_GRAD_NORM:
        axes[3,0].set_title('Grad norm2')
        axes[3,0].set_yscale("log")
        axes[3,0].plot(MEAN_NETSTEP[:,11], color='green',alpha=.5, label='Grad norm2')
        axes[3,0].plot(MEAN_NETSTEP[:,13], color='blue',alpha=.5, label='Accum grad norm2')
        axes[3,0].plot(MEAN_SGD[:,11], color='red',alpha=.5, label='Grad SGD norm2')
        axes[3,0].grid()
        axes[3,0].legend()

        axes[3,1].set_title('Grad norm2 variance')
        axes[3,1].set_yscale("log")
        axes[3,1].plot(MEAN_NETSTEP[:,12], color='green',alpha=.5, label='Grad norm2 variance')
        axes[3,1].plot(MEAN_NETSTEP[:,14], color='blue',alpha=.5, label='Accum grad norm2 variance')
        axes[3,1].plot(MEAN_SGD[:,12], color='red',alpha=.5, label='Grad SGD norm2 variance')
        axes[3,1].grid()
        axes[3,1].legend()

    dummy=fig.suptitle(title)
    plt.show()

def experiments_comparison(data, var, low, high, title="Comparison"):

    MEAN_STRGHT = data[low:high,0]
    MEAN_SGD = data[low:high,1]
    MEAN_NETNTK2 = data[low:high,2]
    MEAN_NETSTEP = data[low:high,3]
    MEAN_NETNTK1 = data[low:high,4]
    if (var is not None):
        low, up = data[low:high]-var[low:high], data[low:high]+var[low:high]
        LOW_STRGHT, UP_STRGHT = low[:,0], up[:,0]
        LOW_SGD, UP_SGD = low[:,1], up[:,1]
        LOW_NETNTK2, UP_NETNTK2 = low[:,2], up[:,2]
        LOW_NETSTEP, UP_NETSTEP = low[:,3], up[:,3]
        LOW_NETNTK1, UP_NETNTK1 = low[:,4], up[:,4]
        cord_x = np.arange(0, len(MEAN_NETNTK2[:,1]))

    display.clear_output()
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    axes[0].set_title('Loss (average cross entropy)')
    axes[0].plot(MEAN_NETNTK2[:,1], color='b',alpha=.5, label='Net-line NTK norm-2')
    if (var is not None):
        axes[0].fill_between(x=cord_x, y1=LOW_NETNTK2[:,1], y2=UP_NETNTK2[:,1], color='b', alpha=.125)
    axes[0].plot(MEAN_NETNTK1[:,1], color='g',alpha=.5, label='Net-line NTK norm-1')
    if (var is not None):
        axes[0].fill_between(x=cord_x, y1=LOW_NETNTK1[:,1], y2=UP_NETNTK1[:,1], color='g', alpha=.125)
    axes[0].plot(MEAN_NETSTEP[:,1], color='b', ls='dotted', alpha=.5, label='Net-line 2-step')
    if (var is not None):
        axes[0].fill_between(x=cord_x, y1=LOW_NETSTEP[:,1], y2=UP_NETSTEP[:,1], color='b', ls='dotted', alpha=.125)
    axes[0].plot(MEAN_STRGHT[:,1], color='orange', alpha=.5, label='Straight-line')
    if (var is not None):
        axes[0].fill_between(x=cord_x, y1=LOW_STRGHT[:,1], y2=UP_STRGHT[:,1], color='orange', alpha=.125)
    axes[0].plot(MEAN_SGD[:,1], color='grey', alpha=.5, label='SGD')
    if (var is not None):
        axes[0].fill_between(x=cord_x, y1=LOW_SGD[:,1], y2=UP_SGD[:,1], color='grey', alpha=.125)
    axes[0].grid()
    axes[0].legend()
    axes[0].set_yscale("log")
    axes[0].set_xlabel("Epoch (series of gradient descent steps)")
    axes[0].set_ylabel("Loss value")

    axes[1].set_title('Error (top 1)')
    axes[1].plot(1.0-MEAN_NETNTK2[:,0], color='b', alpha=.5, label='Net-line NTK norm-2')
    if (var is not None):
        axes[1].fill_between(x=cord_x, y1=1.0-LOW_NETNTK2[:,0], y2=1.0-UP_NETNTK2[:,0], color='b', alpha=.125)
    axes[1].plot(1.0-MEAN_NETNTK1[:,0], color='g', alpha=.5, label='Net-line NTK norm-1')
    if (var is not None):
        axes[1].fill_between(x=cord_x, y1=1.0-LOW_NETNTK1[:,0], y2=1.0-UP_NETNTK1[:,0], color='g', alpha=.125)
    axes[1].plot(1.0-MEAN_NETSTEP[:,0], color='b', ls='dotted', alpha=.5, label='Net-line 2-step')
    if (var is not None):
        axes[1].fill_between(x=cord_x, y1=1.0-LOW_NETSTEP[:,0], y2=1.0-UP_NETSTEP[:,0], color='b', ls='dotted', alpha=.125)
    axes[1].plot(1.0-MEAN_STRGHT[:,0], color='orange', alpha=.5, label='Straight-line')
    if (var is not None):
        axes[1].fill_between(x=cord_x, y1=1.0-LOW_STRGHT[:,0], y2=1.0-UP_STRGHT[:,0], color='orange', alpha=.125)
    axes[1].plot(1.0-MEAN_SGD[:,0], color='grey', alpha=.5, label='SGD')
    if (var is not None):
        axes[1].fill_between(x=cord_x, y1=1.0-LOW_SGD[:,0], y2=1.0-UP_SGD[:,0], color='grey', alpha=.125)
    axes[1].grid()
    axes[1].set_yscale("log")
    axes[1].legend()
    axes[1].set_xlabel("Epoch (series of gradient descent steps)")
    axes[1].set_ylabel("Error value")

    dummy=fig.suptitle(title)
    plt.show()

def rf(value):
    return f"{value:.5f}"

def accuracy_comparison(exp_results, low, high, title="Validation accuracy"):
    strght_mean, strght_var = np.mean(exp_results[:, low:high, 0, 0], axis=0), np.var(exp_results[:, low:high, 0, 0], axis=0)
    sgd_mean, sgd_var = np.mean(exp_results[:, low:high, 1, 0], axis=0), np.var(exp_results[:, low:high, 1, 0], axis=0)
    nline2_mean, nline2_var = np.mean(exp_results[:, low:high, 2, 0], axis=0), np.var(exp_results[:, low:high, 2, 0], axis=0)
    nline_mean, nline_var = np.mean(exp_results[:, low:high, 3, 0], axis=0), np.var(exp_results[:, low:high, 3, 0], axis=0)
    nline1_mean, nline1_var = np.mean(exp_results[:, low:high, 4, 0], axis=0), np.var(exp_results[:, low:high, 4, 0], axis=0)

    p1, p2, p3, p4 = min(5,EPOCHS_PER_EXPERIMENT), min(10,EPOCHS_PER_EXPERIMENT)\
        , min(15,EPOCHS_PER_EXPERIMENT), min(20,EPOCHS_PER_EXPERIMENT)

    report = np.zeros((5,4), dtype='U256')
    for row in range(5):
        mean, var = (strght_mean, strght_var) if row == 0 else (sgd_mean, sgd_var) if row == 1 else \
            (nline2_mean, nline2_var) if row == 2 else (nline_mean, nline_var) if row == 3 else (nline1_mean, nline1_var)
        for col in range(4):
            indx = p1-1 if col == 0 else p2-1 if col == 1 else p3-1 if col == 2 else p4-1
            report[row, col] = rf(mean[indx]) + "+/-" + rf(var[indx])

    fig, ax = plt.subplots(figsize=(8, 8))
    # hide axes
    #fig.patch.set_visible(False)
    ax.axis('off')
    ax.axis('tight')

    tab_plot = ax.table(cellText=report, colLabels=[str(p1)+' epochs',str(p2)+' epochs',str(p3)+' epochs',str(p4)+' epochs']\
        , rowLabels=['Straight-line','SGD','Net-line NTK norm-2', 'Net-line two steps', 'Net-line NTK norm-1'], cellLoc='left', loc='center')
    tab_plot.auto_set_font_size(False)
    tab_plot.set_fontsize(11)
    tab_plot.scale(3, 4)
    #fig.tight_layout()
    ax.set_title(title)
    plt.show()
```

#### Compare optimisers

```python
#Weights distribution variances are set as in (5.67)
cb, cw = 0, 2.0/(slope_plus**2.0 + slope_minus**2.0)
meta = MetaData(batch_size = BATCH_SIZE, output_dim=OUTPUT_DIM, lb = 0.005, lw = 5.0, device=DEVICE)

'''
0-num of experiment; 1-epoch;
2-net: 0-Adam, 1-Net-line 2step, 2-SGD
3-param: 0-test accuracy (Top-1), 1-test errror (cross-entropy), 2-time per step, 3 - qq_norm_test,
4 - pq_norm, 5 - cos_phi, 6 - eta_calculated, 7 - alpha/lr(sgd), 8 - armiho_coeff, 9 - kappa avg, 10 - kappa min,
11 - grad_norm2_sq, 12 - grad_norm2_var, 13 - accum_norm2_sq, 14 - accum_norm2_var, 15 - empty, 16 - empty
'''
experimental_results = np.zeros((EXPERIMENTS, EPOCHS_PER_EXPERIMENT, 5, 15))

for experiment in range(EXPERIMENTS):

    #Straight-line
    net_straightline = MNISTReLU(input_dim=28**2, input_width=4000, hidden_width=1000, output_dim=OUTPUT_DIM)
    net_straightline.set_slopes(slope_plus, slope_minus)
    net_straightline.init_weights(cb, cw)
    net_straightline.to(DEVICE)
    loss_straightline = nn.CrossEntropyLoss(reduction='sum')
    opt_straightline = optim.SGD(net_straightline.parameters(), lr=1e-2)

    #Sgd scheduled lr
    net_sgd = copy.deepcopy(net_straightline)
    net_sgd.set_slopes(slope_plus, slope_minus)
    loss_sgd = nn.CrossEntropyLoss()
    #opt_sgd = optim.SGD(net_sgd.parameters(), lr=1e-2, momentum=0.9, nesterov=True, weight_decay=2.5e-4)
    opt_sgd = optim.SGD(net_sgd.parameters(), 0.02, momentum=0.9)
    schd_sgd = optim.lr_scheduler.CosineAnnealingLR(opt_sgd, T_max=EPOCHS_PER_EXPERIMENT)

    #Net-line NTK norm2
    net_linentk2 = copy.deepcopy(net_straightline)
    net_linentk2.set_slopes(slope_plus, slope_minus)
    opt_linentk2 = StepCalculatorEtaSoftmaxArmihoNorm2Base(meta)

    #Net-line NTK norm1
    net_linentk1 = copy.deepcopy(net_straightline)
    net_linentk1.set_slopes(slope_plus, slope_minus)
    opt_linentk1 = StepCalculatorEtaSoftmaxArmihoNorm1Base(meta)

    #Net-line step
    snl_momentum = 0.9
    snl_eta1 = 0.00001
    snl_foreach = False

    snl_net = copy.deepcopy(net_straightline)
    snl_net.set_slopes(slope_plus, slope_minus)
    snl_opt = optim.SGD(snl_net.parameters(), lr=snl_eta1, momentum=snl_momentum, foreach=snl_foreach)
    snl_sch = NetLineStepLR(snl_net, snl_opt, meta, foreach=snl_foreach)
    snl_sch.do_shorten_lr_for_momentum = True
    snl_sch.do_calc_grad_norm2 = False

    for epoch in range(EPOCHS_PER_EXPERIMENT):
        #alpha for net-line
        snl_sch.alpha_epoch = cosine_annealing2_lr(1.0, 0.0, 0, 20, epoch)

        mt_alpha, mt_agrad, mt_agrad_sq, mt_grad_sga_sq, mt_grad_sga, mt_grad_sq, mt_adam, mt_netline, mt_sgd, mt_qq, mt_pq, mt_cos, mt_eta, mt_armiho, mt_eta_raw, mt_grad = \
            AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter(), AverageMeter()

        for train_batch in train_dataloader:
            images, labels_raw = train_batch
            labels = labels_raw - 1
            images = images.to(DEVICE)
            labels = labels.to(DEVICE)
            batch_size = images.shape[0]
            xx = images.view(batch_size, -1)

            logging.info("####Step with Straight-line")
            logits1 = net_straightline.forward_(xx)
            loss1 = loss_straightline(logits1, labels)
            opt_straightline.zero_grad()
            loss1.backward()
            opt_straightline.param_groups[0]['lr'] = eta_straightline(net_straightline, loss1.item())
            opt_straightline.step()

            logging.info("####Step with SGD-optimiser fixed lr")
            logits2 = net_sgd.forward_(xx)
            loss2 = loss_sgd(logits2, labels)
            opt_sgd.zero_grad()
            loss2.backward()
            #opt_sgd.param_groups[0]['lr'] = lr
            opt_sgd.step()
            # --calculate the gradient--
            sga_grad = 0.
            if CALC_GRAD_NORM:
                for param in net_sgd.parameters():
                    sga_grad += ((param.grad)**2).sum().item()
            mt_grad_sga.update(math.sqrt(sga_grad))
            mt_grad_sga_sq.update(sga_grad)            

            logging.info("####Step with net-line norm2 NTK version")
            with torch.no_grad():
                logits4 = net_linentk2.forward_(xx)
            _, _ = opt_linentk2.step(net_linentk2, logits4.detach().cpu().numpy(), labels, xx)

            logging.info("####Step with net-line norm1 NTK version")
            with torch.no_grad():
                logits5 = net_linentk1.forward_(xx)
            _, _ = opt_linentk1.step(net_linentk1, logits5.detach().cpu().numpy(), labels, xx)

            logging.info("####Step with net-line 2-step version")
            ns_start = time.time_ns()
            step_result = snl_sch.step(labels, xx)
            ns_end = time.time_ns()
            mt_netline.update(ns_end-ns_start)
            mt_pq.update(step_result.pq_norm)
            mt_qq.update(step_result.qq_norm)
            mt_cos.update(step_result.cos_phi)
            mt_eta.update(step_result.eta)
            mt_grad.update(math.sqrt(step_result.grad_norm2_squared))
            mt_grad_sq.update(step_result.grad_norm2_squared)
            mt_agrad.update(math.sqrt(step_result.accum_norm2_squared))
            mt_agrad_sq.update(step_result.accum_norm2_squared)
            mt_alpha.update(step_result.alpha)

        #qq_norm, eta
        experimental_results[experiment, epoch, 3, 3] = mt_qq.avg
        experimental_results[experiment, epoch, 3, 4] = mt_pq.avg
        experimental_results[experiment, epoch, 3, 5] = mt_cos.avg
        experimental_results[experiment, epoch, 3, 6] = mt_eta.avg
        experimental_results[experiment, epoch, 3, 7] = None
        experimental_results[experiment, epoch, 3, 8] = None
        experimental_results[experiment, epoch, 3, 9] = None
        experimental_results[experiment, epoch, 3, 10] = None
        experimental_results[experiment, epoch, 3, 11] = mt_grad.avg
        experimental_results[experiment, epoch, 3, 12] = mt_grad_sq.avg - (mt_grad.avg)**2
        experimental_results[experiment, epoch, 3, 13] = mt_agrad.avg
        experimental_results[experiment, epoch, 3, 14] = mt_agrad_sq.avg - (mt_agrad.avg)**2

        # testing loop
        acc0, loss0 = test_loop(net_straightline)
        experimental_results[experiment, epoch, 0, 0] = acc0
        experimental_results[experiment, epoch, 0, 1] = loss0

        acc1, loss1 = test_loop(net_sgd)
        experimental_results[experiment, epoch, 1, 0] = acc1
        experimental_results[experiment, epoch, 1, 1] = loss1
        experimental_results[experiment, epoch, 1, 6] = schd_sgd.get_last_lr()[0]
        experimental_results[experiment, epoch, 1, 11] = mt_grad_sga.avg
        experimental_results[experiment, epoch, 1, 12] = mt_grad_sga_sq.avg - (mt_grad_sga.avg)**2

        acc2, loss2 = test_loop(net_linentk2)
        experimental_results[experiment, epoch, 2, 0] = acc2
        experimental_results[experiment, epoch, 2, 1] = loss2

        acc3, loss3 = test_loop(snl_net)
        experimental_results[experiment, epoch, 3, 0] = acc3
        experimental_results[experiment, epoch, 3, 1] = loss3

        acc4, loss4 = test_loop(net_linentk1)
        experimental_results[experiment, epoch, 4, 0] = acc4
        experimental_results[experiment, epoch, 4, 1] = loss4

        schd_sgd.step()

        title="Comparison for the same FFN-structure for different optimisers, experiment {}".format(experiment)
        experiment_comparison(experimental_results[experiment, 0:epoch+1], None, 0, 100, title=title)

        print("Experiment={}, net-line={}, straight-line={}, alpha={}"\
            .format(experiment, acc3, acc0, mt_alpha.avg))
```

```python
experimental_mean = np.mean(experimental_results, axis=0)
experimental_var = np.var(experimental_results, axis=0)
title2="Validation on EMINST dataset in {} experiments".format(EXPERIMENTS)
experiments_comparison(experimental_mean, experimental_var, 0, epoch+1, title=title2)
```

```python
title3 = "Validation accuracy mean +/- var on EMINST in {} experiments".format(EXPERIMENTS)
accuracy_comparison(experimental_results, 0, EPOCHS_PER_EXPERIMENT, title=title3)
```
